```python
from pylab import *
from frostsynth.waveform import *
```

```python
t = linspace(-0.1, 1.1, 1000)
s = linspace(-1.1, 1.1, 1000)
```

```python
def multiplot(x, ys):
    args = []
    for y in ys:
        args.extend([x, y])
    plot(*args)
```

```python
plot(t, sine(t))
```

```python
plot(t, cosine(t))
```

```python
plot(t, twine(t))
```

```python
plot(t, halfcircle(t))
```

```python
plot(t, tang(t))
```

```python
plot(t, pinch(t))
```

```python
plot(t, tooth(t))
```

```python
plot(t, tri(t))
```

```python
plot(t, softsaw(t, 0.5))
```

```python
plot(t, softsaw(4 * t, t))
```

```python
plot(t, softarc(4 * t, t))
```

```python
plot(t, tri(4 * t, t))
```

```python
plot(t, tooth(4 * t, t))
```

```python
plot(t, lissajous11(4 * t, s))
```

```python
multiplot(t, [lissajous12(4 * t, s, b) for b in [-0.75, 0, 0.75]])
```

```python
multiplot(t, [lissajous13(4 * t, s, b) for b in [-0.75, 0, 0.75]])
```
