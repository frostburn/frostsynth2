```python
from pylab import *
from frostsynth.oscillator import *
from frostsynth.sampling import *
from frostsynth.ipython import *
from frostsynth.drive import *
from frostsynth.analysis import *
from frostsynth.filter.dynamic import *
from frostsynth.waveform import *
```

```python
set_sample_rate(44100)
```

```python
DUR = 2
t = trange(DUR)
```

```python
freq_driver = lin_driver([1, 2, 1, 3, 1, 4], DUR) * 220
```

```python
plot(t, freq_driver)
```

```python
lpf_driver = lin_driver([4, 1], DUR) * 880
```

```python
plot(t, lpf_driver)
```

```python
signal = sine_blit(freq_driver)
```

```python
Audio(0.01 * signal, autoplay=False)
```

```python
pltspec(signal)
```

```python
filtered = lowpass(sine_blit(freq_driver), lpf_driver, 500)
```

```python
Audio(0.001 * filtered, autoplay=False)
```

```python
pltspec(filtered, 4000)
```

```python
signal = odd_blit(freq_driver * 5)
```

```python
Audio(0.01 * signal, autoplay=False)
```

```python
pltspec(signal)
```

```python
shift = t * 4000
signal = cblit(freq_driver, np.exp(-t*5), shift) * cis(integrate(shift))
signal = signal.imag
```

```python
Audio(0.01 * signal, autoplay=False)
```

```python
pltspec(signal)
```

```python
Audio(signal * 0.5, autoplay=False)
```
