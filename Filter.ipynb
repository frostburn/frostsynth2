```python
from pylab import *
from ipywidgets import *
from frostsynth.filter.dynamic import *
from frostsynth.sampling import *
from frostsynth.analysis import *
from frostsynth.window import cosine
from frostsynth.ipython import Audio
%autoreload
```

```python
set_sample_rate(44100)
t = trange(2)
```

```python
noise = np.random.randn(len(t));filtered = noise
```

```python
def update(filter=Lowpass):
    global filtered
    filtered = filter().map(noise, 2000 + np.sin(10 * t) * 900, 10000)
    imshow(abs(spectrogram(filtered, cosine(1024), 16)), origin="lower")
```

```python
interact_manual(update, filter={"lpf": Lowpass, "hpf": Highpass, "bpf": Bandpass, "notch": Notch, "apf": Allpass})
```

```python
Audio(filtered / abs(filtered).max() * 0.5)
```
